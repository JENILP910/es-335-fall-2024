```python
try:
    from gtts import gTTS # type: ignore
except ImportError:
    %pip install gtts
    from gtts import gTTS
```

```python
def speak(text, file):
    tts = gTTS(text, lang='en')
    with open(file, 'wb') as f:
        tts.write_to_fp(f)
    return Audio(file)
```

```python
speak(transcription['text'], './assets/pikachu.mp3')
```
